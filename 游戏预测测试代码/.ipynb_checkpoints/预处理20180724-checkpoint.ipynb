```python
class skew_dummies(BaseEstimator, TransformerMixin):
    def __init__(self,skew=0.5):
        self.skew = skew
    
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        X_numeric=X.select_dtypes(exclude=["object"])
        skewness = X_numeric.apply(lambda x: skew(x))
        skewness_features = skewness[abs(skewness) >= self.skew].index
        X[skewness_features] = np.log1p(X[skewness_features])
        X = pd.get_dummies(X)
        return X
```

```python
class add_feature(BaseEstimator, TransformerMixin):
    def __init__(self,additional=1):
        self.additional = additional
    
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        if self.additional==1:
            X["abc_gather_level"] = X["sr_rss_c_gather_level"] + X["sr_rss_b_gather_level"] + X["sr_rss_b_prod_level"]   
            X["sr_troop_level"] = X["sr_troop_attack_level"] + X["sr_troop_consumption_level"] + X["sr_troop_defense_level"] + X["bd_hero_strengthen_level"]
            
        else:
            X["abc_gather_level"] = X["sr_rss_c_gather_level"] + X["sr_rss_b_gather_level"] + X["sr_rss_b_prod_level"]   
            X["sr_troop_level"] = X["sr_troop_attack_level"] + X["sr_troop_consumption_level"] + X["sr_troop_defense_level"] + X["bd_hero_strengthen_level"]
            
            X["+sr_gather_level"] = X["sr_rss_e_gather_level"] + X["sr_rss_a_gather_level"]
            X["+sr_speed_level"] = X["sr_healing_speed_level"] + X["sr_alliance_march_speed_level"]+ X['sr_gathering_march_speed_level']
            X["+sr_defense_level"] = X["sr_troop_defense_level"] + X["bd_hero_strengthen_level"]
            X["+sr_hp_level"] = X["sr_infantry_hp_level"] + X["sr_cavalry_hp_level"]
            X["+sr_tier_3_level"] = X["sr_cavalry_tier_3_level"] + X["sr_infantry_tier_3_level"]+X['sr_outpost_tier_3_level']
            X["+sr_tier_4_level"] = X["sr_infantry_tier_4_level"] + X["sr_cavalry_tier_4_level"]
            X["+sr_rss_level"] = X["sr_rss_c_prod_level"] + X["sr_rss_e_gather_level"]
            X["+bd_hal_level"] = X["bd_hall_of_war_level"] + X["bd_hero_strengthen_level"]
            X["+sr_march_level"] = X["sr_march_size_level"] + X["sr_alliance_march_speed_level"]+X['sr_gathering_march_speed_level']
            X["+sr_top_level"] = X["sr_cavalry_tier_4_level"] + X["sr_troop_defense_level"]+X['sr_gathering_march_speed_level']
            
            X["-sr_rss_level"] = X["sr_rss_c_gather_level"] + X["sr_rss_b_gather_level"]+X['sr_rss_b_prod_level']
            X["-sr_tier_2_level"] = X["sr_shaman_tier_2_level"] + X["sr_outpost_tier_2_level"]+X['sr_infantry_tier_2_level']
            X["-acceleration_reduce_value"] = X["reaserch_acceleration_reduce_value"] + X["treatment_acceleration_reduce_value"]+ X["training_acceleration_reduce_value"]+ X["general_acceleration_reduce_value"]
            X["-wound_reduce_value"] = X["wound_shaman_reduce_value"] + X["wound_shaman_add_value"]+ X["wound_infantry_add_value"]
            X["-other_level"] = X["bd_guest_cavern_level"] + X["stone_add_value"]
            X["-sr_shaman_level"] = X["sr_shaman_hp_level"] + X["sr_shaman_tier_2_level"]
            X["-sr_top_level"] = X["sr_rss_c_gather_level"] + X["sr_troop_attack_level"]
           
            X["pay_count_win"] = X["pay_count"] * X["pvp_win_count"]
            X["pay_price_win"] = X["pay_price"] * X["pvp_win_count"]
            X["pay_count_+top_level"] = X["pay_count"] * (X["sr_cavalry_tier_4_level"] + X["sr_troop_defense_level"]+X['sr_gathering_march_speed_level'])
            X["pay_price_+top_level"] = X["pay_price"] * (X["sr_cavalry_tier_4_level"] + X["sr_troop_defense_level"]+X['sr_gathering_march_speed_level'])
            X["pay_count_-top_level"] = X["pay_count"] * (X["sr_rss_c_gather_level"] + X["sr_troop_attack_level"])
            X["pay_price_-top_level"] = X["pay_price"] * (X["sr_rss_c_gather_level"] + X["sr_troop_attack_level"])
            X["pay_count_reduce_value"] = X["pay_count"] * (X["reaserch_acceleration_reduce_value"] + X["treatment_acceleration_reduce_value"]+ X["training_acceleration_reduce_value"]+ X["general_acceleration_reduce_value"])
            X["pay_price_reduce_value"] = X["pay_price"] * (X["reaserch_acceleration_reduce_value"] + X["treatment_acceleration_reduce_value"]+ X["training_acceleration_reduce_value"]+ X["general_acceleration_reduce_value"])
            X["pay_count_mean"] = X["pay_price"] * X["pay_count"]
    
            return X
```

```python
def rmse_cv(model,X,y):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=5))
    return rmse
```

```python
models = [LinearRegression(),Ridge(),Lasso(alpha=0.01,max_iter=10000),RandomForestRegressor(),GradientBoostingRegressor(),SVR(),LinearSVR(),
          ElasticNet(alpha=0.001,max_iter=10000),SGDRegressor(max_iter=1000,tol=1e-3),BayesianRidge(),KernelRidge(alpha=0.6, kernel='polynomial', degree=2, coef0=2.5),
          ExtraTreesRegressor()]
```

```python
names = ["LR", "Ridge", "Lasso", "RF", "GBR", "SVR", "LinSVR", "Ela","SGD","Bay","Ker","Extra"]
for name, model in zip(names, models):
    score = rmse_cv(model, X_scaled, y_log)
    print("{}: {:.6f}, {:.4f}".format(name,score.mean(),score.std()))
```

```python
class grid():
    def __init__(self,model):
        self.model = model
    
    def grid_get(self,X,y,param_grid):
        grid_search = GridSearchCV(self.model,param_grid,cv=5, scoring="neg_mean_squared_error")
        grid_search.fit(X,y)
        print(grid_search.best_params_, np.sqrt(-grid_search.best_score_))
        grid_search.cv_results_['mean_test_score'] = np.sqrt(-grid_search.cv_results_['mean_test_score'])
        print(pd.DataFrame(grid_search.cv_results_)[['params','mean_test_score','std_test_score']])
```

```python
class AverageWeight(BaseEstimator, RegressorMixin):
    def __init__(self,mod,weight):
        self.mod = mod
        self.weight = weight
        
    def fit(self,X,y):
        self.models_ = [clone(x) for x in self.mod]
        for model in self.models_:
            model.fit(X,y)
        return self
    
    def predict(self,X):
        w = list()
        pred = np.array([model.predict(X) for model in self.models_])
        # for every data point, single model prediction times weight, then add them together
        for data in range(pred.shape[1]):
            single = [pred[model,data]*weight for model,weight in zip(range(pred.shape[0]),self.weight)]
            w.append(np.sum(single))
        return w
```

```python
class stacking(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self,mod,meta_model):
        self.mod = mod
        self.meta_model = meta_model
        self.kf = KFold(n_splits=5, random_state=42, shuffle=True)
        
    def fit(self,X,y):
        self.saved_model = [list() for i in self.mod]
        oof_train = np.zeros((X.shape[0], len(self.mod)))
        
        for i,model in enumerate(self.mod):
            for train_index, val_index in self.kf.split(X,y):
                renew_model = clone(model)
                renew_model.fit(X[train_index], y[train_index])
                self.saved_model[i].append(renew_model)
                oof_train[val_index,i] = renew_model.predict(X[val_index])
        
        self.meta_model.fit(oof_train,y)
        return self
    
    def predict(self,X):
        whole_test = np.column_stack([np.column_stack(model.predict(X) for model in single_model).mean(axis=1) 
                                      for single_model in self.saved_model]) 
        return self.meta_model.predict(whole_test)
    
    def get_oof(self,X,y,test_X):
        oof = np.zeros((X.shape[0],len(self.mod)))
        test_single = np.zeros((test_X.shape[0],5))
        test_mean = np.zeros((test_X.shape[0],len(self.mod)))
        for i,model in enumerate(self.mod):
            for j, (train_index,val_index) in enumerate(self.kf.split(X,y)):
                clone_model = clone(model)
                clone_model.fit(X[train_index],y[train_index])
                oof[val_index,i] = clone_model.predict(X[val_index])
                test_single[:,j] = clone_model.predict(test_X)
            test_mean[:,i] = test_single.mean(axis=1)
        return oof, test_mean
```
