```python
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score,roc_curve
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import metrics
import pandas as pd
import numpy as np
import random

data = df_.copy()
data = np.matrix(data)
labels = data[:,108]
labels = np.array(labels)
data = data[:,:108]

no_ones = np.count_nonzero(labels)
no_zeros = labels.size-no_ones

data_threes = np.zeros((1,108))
for record,label in zip(data,labels):
    if(label == 1):
        data_threes = np.append(data_threes, record, axis = 0)
data_threes = np.delete(data_threes, 0, 0)

def smote(samplingAmount, data_threes, data, labels):
    nbrs = NearestNeighbors(n_neighbors=samplingAmount, algorithm='ball_tree').fit(data_threes)
    distances, indices = nbrs.kneighbors(data_threes)
    for record in indices:
        feature = record[0]
        for index in record:
            if(index != feature):
                mul = random.random()
                new = [fv + (mul * (nn - fv)) for nn, fv in zip(data_threes[index], data_threes[feature])]
                data = np.append(data, new[0], axis = 0)
                labels = np.append(labels, 1)
    
    test_data = np.zeros((1,108))
    test_labels = np.zeros(1)
    
    i = 0
    testSet = []
    while i < .1*(data.shape[0]):
        testIndex = random.randint(0,data.shape[0]-1)
        if(testIndex not in testSet):
            testSet.append(testIndex)
            test_data = np.append(test_data, data[testIndex], axis = 0)
            test_labels = np.append(test_labels, labels[testIndex])
            data = np.delete(data, testIndex, 0)
            labels = np.delete(labels, testIndex)
            i = i + 1
    test_data = np.delete(test_data, 0, 0)
    test_labels = np.delete(test_labels, 0)
    print 'smote'
    return data, labels, test_data, test_labels

data1, labels1, test_data1, test_labels1 = smote(3, data_threes[:], data[:], labels[:])
data1_ = pd.DataFrame(data1)#, labels1, test_data1, test_labels1
data1_.columns = df_.columns[:108]
data1_['lable'] = labels1
data1_.to_csv('over_sampling_1.csv',index=None)
```
