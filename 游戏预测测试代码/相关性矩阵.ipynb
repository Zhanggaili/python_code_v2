```python
import pandas as pd

def hebing(data, lable_num_list):
    df = data.iloc[0:,lable_num_list]
    df['Col_sum'] = df.apply(lambda x: x.sum(), axis=1)
    data['hebing'] = df['Col_sum']


data_set = pd.read_csv("youziyuan.csv")
hebing(data_set,[12,14,16])
data_set.rename(index=str, columns={"hebing": "infantry_add_sumvalue"})
# hebing(data_set,[13,15,17])
# data_set.rename(index=str, columns={"hebing": "infantry_reduce_sumvalue"})
# hebing(data_set,[18,20,22])
# data_set.rename(index=str, columns={"hebing": "wound_infantry_add_sumvalue"})
# hebing(data_set,[19,21,23])
# data_set.rename(index=str, columns={"hebing": "wound_infantry_reduce_sumvalue"})
# data_set.drop(columns=data_set.columns[12:24])
print data_set.columns
```
