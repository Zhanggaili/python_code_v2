```python
from numpy import *
import pandas as pd
from sklearn.decomposition import PCA

data_train = pd.read_csv("under_sample_data.csv")
dt1=pd.to_datetime(data_train["register_time"])
data_train["register_time"]=dt1.dt.dayofyear
df = data_train.iloc[0:,1:len(data_train.columns)]

N_components = int(0.9*len(data_train.columns))
pca = PCA(n_components = N_components,copy = True, whiten=True, random_state=0)
df_1 = pca.fit_transform(df)
# print '各方向方差：', pca.explained_variance_
# print '方差所占比例：', pca.explained_variance_ratio_
df_1 = pd.DataFrame(df_1)
df_1['prediction_pay_price'] = df['prediction_pay_price']
df_1.to_csv('under_sample_data_PCA.csv',index=None)
print df_1.head(20)
```
